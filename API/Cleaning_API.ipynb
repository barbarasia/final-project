```python
api = api.drop(columns = ['Category_17', 'Category_16', 'Category_15', 'Category_14', 'Category_13', 'Category_12', 'Category_11', 'Category_10', 'Category_9', 'Category_8', 'Category_7', 'Category_6', 'Countries'])
```

```python
api = api[api['Category_1'] == 'Boissons']
```

```python
api = api[api['Category_2'] == 'Boissons alcoolisées']
```

```python
api = api[api['Category_3'] == 'Vins']
```

```python
# List of categories to drop
categories_to_drop = [
    'Boissons sans alcool', 'Boissons à base de vin', 'Boissons gazeuses', 'Eaux', 
    'Liqueurs', 'Boissons sans sucre ajouté', 'Boissons instantanées', 'Vins de riz'
]

# Filter out rows where 'Category_4' is in the list
api = api[~api['Category_4'].isin(categories_to_drop)]
```

```python
api.isna().sum()
```

```python
api.duplicated().sum()
```

```python
# Remove duplicate rows, keeping the first occurrence
api= api.drop_duplicates()

# Optionally, you can reset the index
api = api.reset_index(drop=True)
```
