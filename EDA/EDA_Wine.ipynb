```python
regions_prod = pd.read_csv("/Users/Barbara/Desktop/Ironhack/Final_Project/final-project/EDA/source_files/Relevé par département de la production des vins (2018).xlsx - JO REC 18.csv")
regions_prod = regions_prod.drop(columns = ['AOP_VSI / VCI', 'TOTAL'])
regions_prod = regions_prod[regions_prod['Departement'] != 'TOTAUX']
regions_prod.rename(columns={'S TOTAL': 'Total_Surface', 'S_AOP': 'Surface_AOP', 'S_IGP': 'Surface_IGP', 'S_VSIG': 'Surface_VSIG'}, inplace=True)

# Replace NaNs with zeros
regions_prod.fillna(0, inplace=True)
```

```python

regions_prod.duplicated().sum()

# Drop rows where all specified columns have zero values
columns_to_check = ['Departement', 'Total_Surface', 'Surface_AOP', 'Surface_IGP',
       'Surface_VSIG', 'AOP_Blanc', 'AOP_Rouge', 'AOP_Rosé', 'IGP_Blanc',
       'IGP_RougeI', 'IGP_Rosé', 'VSIG_Blanc', 'VSIG_Rouge', 'VSIG_Rosé']
regions_prod = regions_prod.loc[~(regions_prod[columns_to_check] == 0).all(axis=1)]



```

```python
regions_prod
```

```python
regions_prod.columns
```

```python
# List of columns to convert
columns_to_convert = ['Total_Surface', 'Surface_AOP', 'Surface_IGP',
       'Surface_VSIG', 'AOP_Blanc', 'AOP_Rouge', 'AOP_Rosé', 'IGP_Blanc',
       'IGP_RougeI', 'IGP_Rosé', 'VSIG_Blanc', 'VSIG_Rouge', 'VSIG_Rosé']  # Add all your column names here

# Remove commas and convert to float for each column
for column in columns_to_convert:
    regions_prod[column] = regions_prod[column].str.replace(',', '').astype(float)
```

```python
regions_prod.dtypes
```

```python
# Melt the dataframe for AOP, IGP, and VSIG production volumes by color
prod_aop = regions_prod.melt(id_vars=['Departement', 'Surface_AOP'], 
                             value_vars=['AOP_Blanc', 'AOP_Rouge', 'AOP_Rosé'], 
                             var_name='Type_Color', value_name='Production_Volume')
prod_igp = regions_prod.melt(id_vars=['Departement', 'Surface_IGP'], 
                             value_vars=['IGP_Blanc', 'IGP_RougeI', 'IGP_Rosé'], 
                             var_name='Type_Color', value_name='Production_Volume')
prod_vsig = regions_prod.melt(id_vars=['Departement', 'Surface_VSIG'], 
                              value_vars=['VSIG_Blanc', 'VSIG_Rouge', 'VSIG_Rosé'], 
                              var_name='Type_Color', value_name='Production_Volume')

# Rename the surface columns
prod_aop = prod_aop.rename(columns={'Surface_AOP': 'Surface'})
prod_igp = prod_igp.rename(columns={'Surface_IGP': 'Surface'})
prod_vsig = prod_vsig.rename(columns={'Surface_VSIG': 'Surface'})

# Add the 'Type' column
prod_aop['Type'] = 'AOP'
prod_igp['Type'] = 'IGP'
prod_vsig['Type'] = 'VSIG'

# Concatenate the dataframes
regions_prod_long = pd.concat([prod_aop, prod_igp, prod_vsig])

# Split the 'Type_Color' column into separate 'Type' and 'Color' columns
regions_prod_long[['Type', 'Color']] = regions_prod_long['Type_Color'].str.extract(r'(\w+)_(\w+)', expand=True)

# Drop the 'Type_Color' column
regions_prod_long.drop(columns=['Type_Color'], inplace=True)

# Arrange the columns in the desired order
long_df = regions_prod_long[['Departement', 'Type', 'Color', 'Production_Volume', 'Surface']]

# Replace 'RougeI' with 'Rouge' in the 'Color' column
long_df['Color'] = long_df['Color'].replace('RougeI', 'Rouge')


# Display the resulting dataframe
long_df['Color'].value_counts()
```

```python
# Ensure 'Departement' column is of string type
regions_prod_long['Departement'] = regions_prod_long['Departement'].astype(str)

# Separate 'Departement' into 'Dept_Code' and 'Dept_Name'
# Correct use of split method
regions_prod_long[['Dept_Code', 'Dept_Name']] = regions_prod_long['Departement'].str.split(n=1, expand=True)
```

```python
regions_prod_long = regions_prod_long.drop(columns=['Departement', 'Dept_Name'])
```
